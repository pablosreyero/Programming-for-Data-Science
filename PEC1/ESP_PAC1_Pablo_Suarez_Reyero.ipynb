   * Sus valores correspondan a cuántos registros contienen este nombre.

2.2 Mostrad las 15 primeras entradas del diccionario (clave y valor), una vez lo hemos ordenado (de mayor a menor) según el número de filas del dataframe que contienen cada valor.

```python
'''Para realizar esta tarea, la única ayuda que he necesitado ha sido la de ordenar el diccionario según las claves. La ayuda la he encontrado en esta pregunta de stackoverflow: https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value'''
```

```python
# Primero se inicializa in diccionario vacío
diccionario_df = {}

#Segundo, se cuenta el número de veces que cada nombre se repite en "First Name":
for nombre in df["First Name"]:
  diccionario_df[nombre] = df["First Name"].value_counts()[nombre]
print(diccionario_df)

#Ahora se comprueba que la suma total del número de veces que aparece cada nombre, coincide con el número de filas del dataframe
print("\n", len(diccionario_df))
print(sum(diccionario_df.values()))
assert len(df) == sum(diccionario_df.values()), "La operación NO se ha realizado correctamente"

#Tercero, se ordenan los elementos del diccionario según el nº de veces aparece cada nombre:
diccionario_df = {clave: valor for clave, valor in sorted(diccionario_df.items(), key=lambda item: item[1], reverse = True)} #Para que ordene las parejas de valores y claves del dicionario de mayor a menor se tiene que poner el flag reverse a True
print("\nEste es el diccionario ordenado de manera ascendente: ", diccionario_df)

#Ahora se vuelve a comprobar que la suma total del número de veces que aparece cada nombre, coincide con el número de filas del dataframe
assert len(df) == sum(diccionario_df.values()), "La operación NO se ha realizado correctamente"
print(len(diccionario_df))

#Ahora se muestran las 15 primeras entradas del diccionario, para eso se hace uso de un bucle.
diccionario_df_quince = {k: diccionario_df[k] for k in list(diccionario_df)[:15]}
print('A continuación se adjuntan las primeras 15 entradas del diccionario; \n',diccionario_df_quince)
```

```python
'''Mi idea era hacer el código un poco interactivo, por eso lo que he hecho ha sido, pedirle al usuario que introduzca el sexo de la persona que quiere inspeccionar, por eso he hecho más código del necesario.'''
```

```python
from pandas.core.arrays import string_
def funcionComprobacion(df: pd.DataFrame, userID: str , sexo: str) -> bool:
  booleano = False

  for iterador, identificador in enumerate(df['User Id']):
    if  identificador == userID:
      nombre_usuario: str = df["First Name"][iterador]
      sexo_usuario: string_ = df["Sex"][iterador]

      if sexo == sexo_usuario:
        booleano: bool = True
      else:
        break

  if booleano:
    print(f"El sexo especificado en el parámetro de entrada para {nombre_usuario} es correcto")
  else:
    print(f"El sexo especificado en el parámetro de entrada para {nombre_usuario} NO es correcto")

  return booleano


id = "33730caFF13Ff4F"
sexos = ["Female","Male"]

while True:
    sexo: str = str(input("Please specify user's sex :\n"))
    if sexo in sexos:
      booleano: bool = funcionComprobacion(df, id, sexo)
      break
    else:
      print("Invalid, please indicate wether the user you want to check is either <<Male>> or <<Female>>\n")
      continue
```

```python
#Test
assert funcionComprobacion(df, id, sexo) == True, "El sexo introducido no coincide con el de la base de datos."
```

```python
#Comentarios y referencias:

'''Para el punto 5.3, para contar el número de conocimientos, he creído más fácil contar el número de comas, ya que si cada conocimiento está separado por el número de comas, entonces simplemente habría que contar el número de comas y sumarle un 1, pues: #_comas = #_strings - 1. Creo que a lo mejor está sería la única manera por el hecho de que no se podría realizar la implementación contando strings, porque gran parte de los conocimientos están formados por dos palabras.
En este caso, yo obtengo un diccionario de 690 pares de elementos, lo que significa, que en este caso, no hay nombres repetidos, por lo tanto el número de conocimientos de cada nombre, se corresponde con el número de conocimientos del último registro relativo a ese nombre. Además con este procedimiento se ha verificado el tamaño del diccionario que se ha creado en el ejercicio 2.
No obstante, se ha realizado el mismo procedimiento, pero en este caso para el userID, haciendo esto, nos saldrá un tamaño de diccionario equivalente al número del filas del dataframe porque los userID de cada persona son únicos y no se repiten.
Para buscar el valor máximo dentro de un diccionario, me he ayudado del siguiente ejemplo en stackoverflow: https://stackoverflow.com/questions/268272/getting-key-with-maximum-value-in-dictionary
Para la última parte, para poder obtener el índice del nombre de la persona con más conocimientos (ya que el anterior apartado lo hice con los Users Ids porque así no se repetían) he tenido que buscar en stackoverflow un error que me daba acerca de que los diccionarios no tenían la función .index, por eso he puesto "list(nuevo_diccionario_userid.keys()).index(max_num_conocimientos[0])" con list, para poder obtener el índice.
El link a la página es el siguiente: https://stackoverflow.com/questions/14538885/how-to-get-the-index-with-the-key-in-a-dictionary
'''
```

```python
# 5.1º) Cread una lista con los empleos "Job Title" de las personas sin repeticiones.
lista_de_empleos: list = []
for i in df["Job Title"]:
  if i not in lista_de_empleos:
    lista_de_empleos.append(i)

assert len(df) != len(lista_de_empleos), "Hay repeticiones de empleos en la lista que se acaba de crear"

# 5.2º) Ordenadla (alfabéticamente, el orden establecido por Python es ya suficidente) y mostrad los 10 primeros valores.
lista_de_empleos = sorted(lista_de_empleos)
diez_primeros_valores = [i for i in lista_de_empleos[:10]]
print("\nEstos son los 10 primeros empleos de la lista ordenada alfabéticamente: \n",diez_primeros_valores)

# 5.3º) Cread un diccionario (se valorará el uso de Dict comprenhension) tal que, las claves correspondan al First Name y los valores correspondan al número de conocimientos que contiene el "Job title". Consideramos que una persona tiene varios conocimientos en el "Job Title", separados por comas.
nuevo_diccionario = {k: v.count(",")+1 for k,v in zip(df["First Name"],df['Job Title'])}
print("\nEste es el nuevo diccionario: \n",nuevo_diccionario)
print("Este es el tamaño del diccionario: ",len(nuevo_diccionario))

nuevo_diccionario_userid = {k: v.count(",")+1 for k,v in zip(df["User Id"],df['Job Title'])}
print("\nEste es el nuevo diccionario dónde las claves son los IDs, ya que si se hace por nombres el resultado es el de arriba y es erróneo: \n",nuevo_diccionario_userid)
print("Este es el tamaño del diccionario: ",len(nuevo_diccionario_userid))

# 5.4º) Del apartado anterior mostrad debidamente formateado, cuál es el nombre de persona (First Name) que tiene un número mayor de conocimientos y cuántos conocimientos tiene.
#Como en el diccionario que recoge el número de conocimientos por nombre, almacena para cada nombre el número de conocimientos del útlimo registro con ese nombre
#lo que se va a hacer es comprobar que el número de IDs no se repite, por eso, se va a crear una lista de los User Ids únicos, y su tamaño se comparará con el tamaño del df (número de filas)
#si estos dos tamaños son iguales, entonces este apartado se va a realizar con el diccionario del userid y a partir de la posición en el diccionario se podrá extraer el nombre de la persona que tiene el mayor
#número de conocimientos
lista_ids = []
for i in df["User Id"]:
  if i not in lista_ids:
    lista_ids.append(i)

print("\n",len(lista_ids),len(df["User Id"])) #Como se puede ver es el mismo tamaño, por lo tanto se usará el diccionario con los User Ids, ya que aunque nombres repetidos en el dataframe, se tratan de personas distintas.
assert len(lista_ids) == len(df["User Id"]), "El tamaño NO es el mismo"

max_num_conocimientos = max(nuevo_diccionario_userid.items(), key = lambda k:k[1]) #Se devuelve una lista con la clave y su valor (#_conocimientos)
print(max_num_conocimientos) #(user_id, número de conocimientos)
indice_nombre = list(nuevo_diccionario_userid.keys()).index(max_num_conocimientos[0])
nombre_con_mas_conocimientos = df["First Name"][indice_nombre]
apellido_con_mas_conocimientos = df["Last Name"][indice_nombre]
print(f"\nLa persona con más conocimientos es {nombre_con_mas_conocimientos} {apellido_con_mas_conocimientos} con {max_num_conocimientos[1]} conocimientos")


```
