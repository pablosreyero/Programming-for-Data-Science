**Solución**

Como en esta tarea solo se puede hacer uso del material de la asigntura, la solución que se ha propuesto, hace uso de la recursividad. Esta es la única forma de calcular el tamaño de un directorio que alberga un número indeterminado de carpetas. La recursividad consiste en llamar a la función dentro de ella misma. En este caso, esto nos sirve para poder saber si los elementos que está leyendo, son directorios o ficheros, y consecuentemente, actuar de una manera u otra, dependiendo del caso.


````python
import os
import zipfile as zf
import hashlib

# Respuesta al ejercicio 2.1


def check_file_size(directorio):

    """Esta función devuelve el tamaño del directorio o archivo especificado.

      Esta función acepta un único parámetro de tipo string, correspondiente
      a la ruta que se quiere inspeccionar.

      Argumento de entrada:
        -directorio -- ruta especificada (type = string)

      Variable de retorno:
        -tamano_final: tamaño total del/los elemento·s
    """

    if os.path.isfile(directorio):
        file_size_archivo = os.path.getsize(directorio)
        file_size_archivo = file_size_archivo/(1024**2)
        return file_size_archivo

    else:
        file_size = 0
        elementos = os.listdir(directorio)
        for elemento in elementos:
            directorio_elemento = os.path.join(directorio, elemento)
            file_size += check_file_size(directorio_elemento)
        return file_size

    return 0


# directorio = "/content/drive/MyDrive/Colab_Notebooks/activity_2/original.zip"
directorio = "./original.zip"
size = check_file_size(directorio)

print("El tamaño del fichero en: {} es {} MB\n".format(directorio, size))

# Ahora se comienza con la parte de la decompresión;
print("\n\n Ahora se descomprime el archivo \"original.zip\"")


with zf.ZipFile(directorio, 'r') as zip_f:
    # Mostramos el contenido del zip
    # print(zip_f.printdir())
    # Descomprimimos todo el contenido del zip
    zip_f.extractall("/content")

# Primero cojo la ruta del archivo descomprimido:
directorio_script = os.path.dirname(os.path.abspath("original"))

# Ahora le adjunto el nombre del archivo descomprimido, para luego acceder
# a el correctamente en la función que se ha diseñado antes:
directorio_completo = os.path.join(directorio_script, "original")
print(directorio_completo)

# Se calcula a continuación el tamaño del fichero descomprimido.
size_decompressed = check_file_size(directorio_completo)

print("El tamaño del fichero en: {} es {} MB".format(directorio_completo,
                                                     size_decompressed))
````

````python
def struct_a_diccionario(ruta):

    """Esta función devuelve el diccionario con el contenido demandado.

      Esta función acepta un único parámetro de tipo string, correspondiente
      a la ruta que se quiere inspeccionar.

      Argumento de entrada:
        -directorio: ruta especificada (type = string)
      Variable de retorno:
        -diccionario: tamaño total del/los elemento·s
    """

    lista_dicts = []
    diccionario_medio = {}
    clau = 0
    with os.scandir(ruta) as estados:
        for estado in estados:
            ruta_estados = os.path.join(estado)
            with os.scandir(ruta_estados) as ciudades:
                for ciudad in ciudades:
                    ruta_ciu = os.path.join(ciudad)
                    with os.scandir(ruta_ciu) as dir_list:
                        for entry in dir_list:
                            if entry.is_file() and entry.name.endswith(".txt"):
                                ruta_txt = os.path.join(ruta_ciu, entry.name)
                                lista_dicts.append(ruta_txt)
                                with open(ruta_txt, 'r') as file:
                                    diccionario = {}
                                    for linea in file:
                                        diccionario[clau] = {}
                                        diccionario[clau]['summary'] = linea.split(';')[0]
                                        diccionario[clau]['city'] = os.path.basename(
                                            os.path.dirname(ruta_txt))
                                        diccionario[clau]['state'] = os.path.basename(
                                            os.path.dirname(ruta_ciu))
                                        diccionario[clau]['date_time'] = linea.split(';')[1]
                                        diccionario[clau]['shape'] = os.path.splitext(entry.name)[0]
                                        diccionario[clau]['duration'] = linea.split(';')[2]
                                        diccionario[clau]['text'] = linea.split(';')[3]
                                        diccionario[clau]['city_latitude'] = linea.split(';')[4]
                                        diccionario[clau]['city_longitude'] = linea.split(';')[5].strip()
                                        diccionario_medio = diccionario_medio | diccionario
                                        clau += 1
    # La función strip lo que hace es eliminar el simbolo del salto de linea
    print('Este es el tamaño del diccionario: ', len(diccionario_medio))
    print('Este es el número de archivos .txt totales: ', len(lista_dicts))
    return diccionario_medio


ruta = './original'
myDict = struct_a_diccionario(ruta)
# help(struct_a_diccionario)
````

Este código lo he implementado para comprobar el diseño del diccionario que tenía que hacer, es decir, juntando todas las lineas de los archivos txt. Es solo un código de comprobación.

````python
import os


def contar_carpetas(directorio):

    contador = 0
    arxius_txt = []
    total_lineas = 0
    # Recorre recursivamente los subdirectorios
    for raiz, directorios, archivos in os.walk(directorio):
        # Añade las carpetas a la cuenta
        contador += len(directorios)

        elementos = os.listdir(raiz)
        # Filtra solo los archivos con extensión .txt
        for elemento in elementos:
            if elemento.endswith('.txt'):
                arxius_txt.append(elemento)
                ruta_archivo = os.path.join(raiz, elemento)
                with open(ruta_archivo, 'r') as archivo:
                    lineas = archivo.readlines()
                    total_lineas += len(lineas)
    return contador, len(arxius_txt), total_lineas


# Reemplaza 'tu_ruta' con la ruta del directorio que quieres analizar
directorioo = './original'

num_carpetas, num_txt, total_lineas = contar_carpetas(directorioo)
print(f"Número de carpetas en '{directorioo}' y subdirs: {num_carpetas}")
print(f"Número de archivos txt en '{directorioo}': {num_txt}")
print(f"Número de lineas en '{directorioo}': {total_lineas}")
````

````python
import re
import pandas as pd
from itertools import islice
import zipfile as zf

# Esta función la ha creado para poder iterar en el diccionario y comprobarlo


def take(n, iterable):

    """Esta función devuelve los primeros n términos.

      La función acepta dos argumentos, correspondientes a los n primeros
      elementos que se quieren obtener del iterable, y el diccionario que se
      quiere recorrer .

      Argumentos de entrada:
        - n -- "n" primeras posiciones, i.e., type(n) = 'int'
        - iterable -- un diccionario, i.e., type(iterable) = 'dict'

      Variable de retorno:
        - Los "n" primeros elementos del iterable en forma de lista

     """
    return list(islice(iterable, n))

# Esta función es la relativa al apartado 3.4


def dict_and_dir(ruta, df):

    """
    Esta función crea un diccionario que cumple con las instrucciones del
    apartado 3.4, a partir de un df y lo devuelve. Además, esta función también
    se encarga de crear una estructura de ficheros a partir del diccionario
    creado antes, y que cumple con el punto 2 del apartado 3.4.

      La función acepta dos argumentos, correspondientes a los n primeros
      elementos que se quieren obtener del iterable, y el diccionario que se
      quiere recorrer .

      Argumentos de entrada:
        - ruta -- ruta donde se crea el sistema de ficheros, type = 'str'
        - df -- dataframe, type = 'pandas'

      Variable de retorno:
        - Los "n" primeros elementos del iterable en forma de lista
     """

    # Para poder pasarlo todo a un diccionario como antes, se ha de añadir una
    # columna con los indices, pues los indices en el dataframe no tienen
    # columa dedicada asi, se obtendrá el mismo diccionario que aquel creado en
    # el ejercicio 2.2.

    # Primero se crea un diccionario a partir del DF
    diccionario = df.set_index(df.index).to_dict(orient='index')
    n_items = take(10, diccionario.items())
    print(n_items[:10])

    # Ahora se crea el sistema de directorios, recorriendo el diccionario
    # siguiendo las instrucciones del enunciado

    for evento in diccionario.items():
        summ = evento[1]['summary']
        city = evento[1]['city']
        state = evento[1]['state']
        time = evento[1]['date_time']
        shape = evento[1]['shape']
        dur = evento[1]['duration']
        # text = evento[1]['text']
        latitude = evento[1]['city_latitude']
        longitude = evento[1]['city_longitude']
        year = evento[1]['year']

        # Directorio raíz
        year_dir = os.path.join(ruta, year)
        os.makedirs(year_dir, exist_ok=True)

        # Directorio de estados
        state_dir = os.path.join(year_dir, state)
        os.makedirs(state_dir, exist_ok=True)

        # Directorio de ciudades
        city_dir = os.path.join(state_dir, city)
        os.makedirs(city_dir, exist_ok=True)

        # Archivo de texto para los evento
        archivo_objeto = shape+'txt'
        ruta_archivo = os.path.join(city_dir, archivo_objeto)

        # Crear o agregar contenido al archivo
        with open(ruta_archivo, 'w') as file: # En modo 'a' para sobre-escribir
            file.write(f"{summ};{time};{dur};{latitude};{longitude}\n")
    return diccionario


# SOLUCIÓN al apartado 3.1
df['year'] = [year[:4] for year in df['date_time']]

# SOLUCIÓN al apartado 3.2
# Se definen las dos expresiones regulares, para buscar según el criterio
reg1 = r'\bE[a-z]*\b \bT[a-z]*\b'
reg2 = r'\b[uU][a-zA-Z]*\b \b[fF][a-zA-Z]*\b \b[oO][a-zA-Z]*\b'

# Ahora se inicializa el data frame nuevo: df32
df32 = pd.DataFrame()

for subtexto in df['text']:
    if re.findall(reg1, subtexto) and re.findall(reg2, subtexto):
        x = re.findall(reg1, subtexto)
        y = re.findall(reg2, subtexto)
        df32 = pd.concat([df32, df[df['text'] == subtexto]],
                         ignore_index = True)

print(df32.head(len(df32)-1))

# SOLUCIÓN al apartado 3.3
df33 = df.drop('text', axis = 1)

df33.head(len(df33)-1)

# SOLUCIÓN al apartado 3.4
ruta = './root_folder2'
diccionario = dict_and_dir(ruta, df)

# SOLUCIÓN al apartado 3.5
ruta1 = './root31'
ruta2 = './root32'
ruta3 = './root33'

diccionario1 = dict_and_dir(ruta1, df)
diccionario2 = dict_and_dir(ruta2, df32)
diccionario3 = dict_and_dir(ruta3, df33)

# Ahora se calculan los tamaños:
size_31 = check_file_size(ruta1)
size_32 = check_file_size(ruta2)
size_33 = check_file_size(ruta3)

# Ahora se muestra por pantalla los tamaños de los tres directorios
print(f'\nEl directorio {ruta1} pesa: {size_31} MB')
print(f'El directorio {ruta2} pesa: {size_32} MB')
print(f'El directorio {ruta3} pesa: {size_33} MB')
````
