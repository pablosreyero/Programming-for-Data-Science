## Ejercicio 1
El número $\pi$ se puede aproximar con la siguiente serie de sumandos:

    pi = 4 (1-1/3+1/5-1/5+1/7-1/7+1/9...)

sólo hace falta dar más valores a la serie, y obtendremos una mejor aproximación. Esta serie la podemos implementar con la siguiente función:

    def approximate_pi(N):
        value = 0
        for k in range(1, 2*N+1, 2):
            sign = -(k % 4 - 2)
            value += float(sign) / k
        return 4 * value

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">SM</span> **(1 punto)**

### 1.1
Cuenta los pasos que realiza la función **approximate_pi()** y deduce su complejidad asintótica en el peor caso (notación $\mathcal{O}$), a partir del número de pasos que se ejecutan.

**Solución 1.1**.

Los pasos llevados a cabo por la función `def (approximate_pi(N))` son los siguientes:

* Se inicializa la variable donde se guardará la aproximación a 0.
* Recorre una lista $l$ de la forma: $l = \{1, 2, 3, ..., 2 N+1 \}$ tomando en orden todos los valores de dos en dos, que resultan ser los números impares (nones) de la misma, i.e., $(2 \alpha+1)$ $\in$ $l$  $\hspace{3mm}\forall \alpha$ $\in$ $\{0, 1, 2, ..., N\}$
* Calcula la operación módulo entre k y 4, se le restan dos unidades a la operación anterior, y se le cambia el signo, a fin de obtener una suma ponderada.
* Se transforma el resutado anterior de tipo `int` a `float`, se divide la nueva variable entre el número k y por último, al valor actual de `value` se le suma su anterior valor.
* Se multiplica el resultado anterior por 4 y se retorna como valor.

Sabiendo los pasos, ya estamos en condiciones de poder estimar el tiempo de ejecución del algoritmo en el peor de los casos, en función de $N$. El peor de los casos para este algoritmo, es aquel, en el cual el bucle se ejecuta para el valor máximo de N. Esto significa que se realizarán más operaciones para calcular la suma, puesto que cuanto mayor sea N, más términos añadiremos a la aproximación de $\pi$. Nótese como a partir de ahora $n=N$, por lo que formalmente podríamos decir, que $\exists t_w$, tal que $t_w : n \rightarrow \mathbb{R}$, dónde $t_w$ determina el tiempo de ejecución del algoritmo del peor caso. Analizando el algoritmo, la expresión a la que se llega de $t_w$, es la siguiente: 

$$
t_w(n) = 1 + n + n + n + 1 \equiv 2 + 3 n
$$

* El primer $1$ se corresponde con la inicialización de la variable `value`
* La primera $n$ se corresponde con la recuperación de cada uno de los valores de la lista, i.e., `for k in range(1, 2*N+1, 2)`
* La segunda $n$ se corresponde con el cálculo de la variable `sign = -(k % 4 - 2)`
* La tercera $n$ se corresponde con el cálculo de la variable `value += float(sign)/k`
* El último $1$ se corresponde con el retorno de la variable.

Conociendo los pasos que da el algoritmo a fin de poder aproximar el valor de $\pi$, así como la expresión del tiempo de ejecución en el peor caso, asumiendo que todas las operaciones duran lo mismo (aunque en la práctica esto no sea así) ya se puede estimar la complejidad asintótica con la notación $\mathcal{O}(·)$. 

No obstante, antes de mencionar en términos de $\mathcal{O}(·)$ la complejidad del algoritmo, vamos a seguir un poco con la dinámica expuesta en el *notebook* *Optimització_complexitat_i_profiling*. Por ello, observando la expresión de arriba, se puden inferir dos aspectos fundamentales:

* La función $t_w(N)$ está formada por dos términos: $2$ y $3n$. Por ello, si uno observa el comportamiento de la función cuando $n$ crece notablemente (es decir, si observamos el **comportamiento asintótico** de la función), el primer término es completamente despreciable, ya que el segundo término tiene mucho más peso. De este modo, uno puede simplificar la expresión del tiempo de ejecución del algoritmo manteniendo únicamente el término de más peso, $3n.$ Por ello, podríamos decir que según $\uparrow n$ ello implica que $t_w(n)\approx 3n$.
* Aplicando el mismo razonamiento, se tendría que, cuando el tamaño de la entrada crece, uno podría simplificar un poco más la expresión, y suprimir el factor $3$ que multiplica a $n$, por lo que la expresión de $t_w(n)$ quedaría como sigue:  $t_w (n)=n$.

Dicho lo anterior, podemos decir entonces que, $2 + 3n \in \Theta(n)$. Ahora bien, porque dada una función $g(n)$, se denota con $\Theta(g(n))$ el siguiente conjunto de funciones:

$$\Theta(g(n)) = \{ f(n) : 
    \exists c_1, c_2 \text{ y } n_0 > 0
    \text{ tales que  }
    0 \leq c_1g(n) \leq f(n) \leq c_2g(n),
    \forall n \ge n_0 \}$$

Es decir, $\Theta(g(n))$ es el conjunto de funciones para los que existen constantes positivas $c_1$ y $c_2$ tales que la función se encuentra *en medio* de $c_1g(n)$ y $c_2g(n)$ para valores $n$ grandes (valores de $n$ superiores a $n_0$).

Ahora, hacemos atención a la notación que se nos pide en esta pregunta, i.e., $\mathcal{O}(·)$, que resulta especialmente útil, cuando uno conoce solo los límites superiores. Dada una función $g(n)$, se denota con $\mathcal{O}(g(n))$ el conjunto de funciones siguiente:

$$\mathcal{O}(g(n)) = \{ f(n) : 
    \exists c_2 \text{ y } n_0 > 0
    \text{ tales que  }
    0 \leq f(n) \leq c_2g(n),
    \forall n \ge n_0 \}$$

Volviendo a la definición de $\Theta$, la notación $\mathcal{O}$, resulta muy similar a la misma, pero, en esta última, la función solo queda limitada por arriba (se han reducido las restricciones en cuanto a los valores mínimos de la función).

Conociendo las dos definiciones y observando la expresión del tiempo de ejecución para el peor de los casos, i.e., $t_w = 2 + 3n$ podemos decir con seguridad que la complejidad del algoritmo expuesto en este ejercicio puede enunciarse con $\mathcal{O}(n)$, puesto que (entre otros) $t_w (n) \propto n$.

```python
def approximate_pi(N):
    'Esta función aproxima el valor de pi'
    value = 0
    for k in range(1, 2*N+1, 2):
        sign = -(k % 4 - 2)
        value += float(sign) / k
    return 4 * value


approximate_pi(N=10)
```

**Solución 1.2**

Habiendo deducido la complejidad del algoritmo matemáticamete arriba, se va a proceder a demostrar como aquello dicho antes, es cierto. Para acometer esta tarea se ha implementado el siguiente código, producto de la inspiración a partir de uno de los códigos del *notebook*: *Optimització_complexitat_i_profiling*.

```python
def plot_funs_theta(xs, n_val=False, sup_only=False):
    """
    Plot the function 2+3n together with a lower bound (2x) and an
    upper bound (4x).
    """
    legend, c1, c2, n0 = [], 2, 4, 1

    ys_f, leg_f = [2 + 3*n for n in xs], "$2+3n$"
    ys_low, leg_low = [c1*n for n in xs], "$2n$"
    ys_high, leg_high = [c2*n for n in xs], "$4n$"

    fig = plt.figure()

    plt.plot(xs, ys_f, '-', color='tab:orange')
    legend.append(leg_f)
    plt.plot(xs, ys_high, ':', color='tab:green')
    legend.append(leg_high)

    if not sup_only:
        plt.plot(xs, ys_low, ':', color='tab:blue')
        legend.append(leg_low)

    if n_val:
        plt.plot([n0, n0], [0, max(ys_high)], '--', color='tab:gray')

    plt.legend(legend)
    plt.xlabel('$n$')
    plt.ylabel('$f(n)$')
    plt.title('Function growth')
    plt.show()


# Ahora llamámos a la función de arriba
plot_funs_theta(range(10000000), False, True)
```

Se puede comprobar, como efectivamente, la complejidad del algoritmo puede ser expresada como $\mathcal{O}(n)$ pues se trata de un crecimiento lineal.

**Solución 3.2**

El método de búsqueda Sentinel, como se ha comentado en la descripción de su función arriba, es una variante del método *linear_search*. Es decir, lleva a cabo una búsqueda lineal para encontrar un valor específico, pero a diferencia del método *linear_search*, el método *sentinel_search* sustituye el último valor de la lista, por un valor centinela que se corresponde con el valor que se quiere encontrar. Los pasos que implementa este algoritmo son los siguientes. Primero se inicializa el índice de búsqueda a 0, luego se establece el último elemento de la lista como el centinela. Ahora bien, mientras que el centinela no sea igual al elemento actual de la lista que se está recorriendo, el índice de búsqueda se incrementa. Si el índice de búsqueda es menor que el tamaño de la lista o, si el elemento correspondiente al índice de búsqueda en la lista es igual al centinela, se devuelve el valor del índice donde se encuentra el valor que se quería encontrar. Pero, en caso contrario, el centinela no se encuentra en la lista por lo que se devuelve `None`.


En el caso de este algoritmo solo hay que verificar si el valor actual es diferente al centinela, por lo que no hay que verificar el valor deseado. Esto tiene una repercusión en la rápidez de búsqueda, ya que elimina la necesidad de una comparación adicional y consecuentemente afecta de manera positiva al rendimiento de la búsqueda.

Aunque este algoritmo puede mejorar el tiempo de ejecución en un caso promedio, la complejidad respecto al tiempo de ejecución en el peor de los casos (al igual que su predecesor (*linear_search*)), en notación $\mathcal{O}$ se expresa como $\mathcal{O}(n)$. Por lo que el *sentinel_search* no mejora el tiempo de ejecución respecto al *linear_search* en el peor caso.

```python
def linear_search(arr, N, x):
    """
    Esta función acomete una búsqueda lineal, dónde:

    -Cada elemento se considera una posible coincidencia
    con la clave y se comprueba si lo es.

    -Si se encuentra elemento igual a la clave,
    la búsqueda tiene éxito y se devuelve su índice.

    -Si no se encuentra ningún elemento igual a la clave,
    la búsqueda arroja el siguiente resultado:
        "No se ha encontrado ninguna coincidencia".

    Los parámetros de entrada de esta función, son:
        -arr: La lista de elementos.
        -x: El nº a encontrar.
        -N: El tamaño de la lista.
    Cuando la función termina encuentra el nº lo devuelve
    junto con el tiempo transcurrido. En caso de que el nº
    que se quiere buscar no existiese, entonces, la función
    devuelve "-1".
    """
    start_time = time.time()
    for i in range(0, N):
        if (arr[i] == x):
            # print(f'El nº era {x}, y se ha encontrado: {arr[i]} (lin)')
            stop_time = time.time()
            lin_time = stop_time - start_time
            return lin_time
    return -1


def sentinel_search(arr, N, x):
    """
    Esta función es una variante de la función anterior, dónde:

    -Inicializa la variable de índice de búsqueda i a 0.

    -Establece el último elemento del array como la clave de búsqueda.

    -Mientras la clave de búsqueda no sea igual al elemento actual
    del array (es decir, arr[i]), incrementa el índice de búsqueda i.

    -Si i es menor que el tamaño de la matriz o arr[i] es igual a la
    clave de búsqueda, devuelve el valor de i.

    -En caso contrario, la clave de búsqueda no está presente
    en la matriz, por lo que devuelve -1 (o cualquier otro valor
    apropiado para indicar que no se ha encontrado la clave).

    Los parámetros de entrada de esta función, son:
        -arr: La lista de elementos.
        -x: El nº a encontrar.
        -N: El tamaño de la lista.

    Cuando la función termina, esta devuelve "-1".
    """
    # comienza a correr el tiempo
    start_time = time.time()

    # Último elemento de la lista
    last = arr[N - 1]

    # El elemento a encontrar
    # se introduce en la última posición
    arr[N - 1] = x
    i = 0

    while (arr[i] != x):
        i += 1

    # Reponer el último elemento
    arr[N - 1] = last

    if ((i < N - 1) or (arr[N - 1] == x)):
        # print(f'El nº era {x}, y se ha encontrado: {arr[i]} (sent)\n')
        stop_time = time.time()
        sent_time = stop_time - start_time
        return sent_time


def figuras(lista_equi, times_lin, times_sent):
    """
    Esta función recibe como argumentos, todos los tiempos
    de ejecución, que son los siguientes:
        -times_lin: vector con los tiempos de las ejecuciones
        acometidas por el algoritmo `linear_search`
        -times_sent: vector con los tiempos de las ejecuciones
        acometidas por el algoritmo `sentinel_search`

    Nótese como esta función no devuelve ningún parámetro o
    valor, simplemente se limita a gráficar los tiempos de
    ejecución de las dos funciones.
    """
    # print(f'Este es el tamaño de times_lin: {len(times_lin)}')
    # print(f'Este es el tamaño de times_sent: {len(times_sent)}')

    # print(f'Este es el primer tamaño de times_lin: {len(times_lin[1])}')
    # print(f'Este es el primer tamaño de times_sent: {len(times_sent[1])}')

    plt.figure()
    plt.plot(lista_equi, times_lin, marker='o')
    plt.plot(lista_equi, times_sent, marker='o')
    plt.legend(["linear_search", "sentinel_search", "función lineal ejemplo"])
    plt.xlabel('$n$')
    plt.ylabel('$t(n)$')
    # plt.xscale('log')
    # plt.yscale('log')
    plt.title("Tiempos de ejecución para ambos tipos de búsqueda")
    plt.show()


if __name__ == "__main__":
    # Creamos la lista de los 100 elems equiespaciados
    # lista_equi = [10, 100, 1000, 10000, 100000, 1000000]
    lista_equi = [x for x in range(11, 111)]
    print(f"Lista generada: \n {lista_equi}\n")

    # Ahora se definen los tiempos
    times_lin = []
    times_sent = []

    for i in lista_equi:
        lista_i = lst_bloques[:i]
        # print(f'Esta es la lista {i}\n{lista_i}')
        tam = len(lista_i)

        last_el = lista_i[-1]
        # print(f'{last_el} es el de la lista {i}')

        lin_time = linear_search(lista_i, tam, last_el)
        sent_time = sentinel_search(lista_i, tam, last_el)

        # Agregamos los tiempos de exec para cada tamaño
        times_lin.append(lin_time)
        times_sent.append(sent_time)

    print(f'Los tiempos de linear_search(): \n{times_lin}\n')
    print(f'Los tiempos de sentinel_search(): \n{times_sent}')

    # print(f'Este es el tamaño de la lista_equi: {len(lista_equi)}')
    # print(f'Este es el tamaño de times_lin: {len(times_lin)}')
    # print(f'Este es el tamaño de times_sent: {len(times_sent)}')

    # representamos entonces
    figuras(lista_equi, times_lin, times_sent)
```

```python
def binary_search(arr, l, r, x): # iterative approach
    """
    La búsqueda binaria se define como un algoritmo de búsqueda
    utilizado en una matriz ordenada dividiendo repetidamente el
    intervalo de búsqueda por la mitad. La idea de la búsqueda
    binaria es utilizar la información de que la matriz está ordenada
    y reducir la complejidad temporal a O(log N).

    Parámetros de entrada:
        -arr: es la lista donde se va a llevar a cabo la búsqueda.
        -l: es el índice donde se va a comenzar a buscar el nº.
        -r: es el tamaño de la lista.
        -x: es el nº que se quiere buscar en la lista.
    """
    while l <= r:

        mid = l + (r - l) // 2

        # Check if x is present at mid
        if arr[mid] == x:
            return mid

        # If x is greater, ignore left half
        elif arr[mid] < x:
            l = mid + 1

        # If x is smaller, ignore right half
        else:
            r = mid - 1

    # If we reach here, then the element
    # was not present
    return -1


def linear_search(arr, N, x):
    """
    Esta función acomete una búsqueda lineal, dónde:

    -Cada elemento se considera una posible coincidencia
    con la clave y se comprueba si lo es.

    -Si se encuentra elemento igual a la clave,
    la búsqueda tiene éxito y se devuelve su índice.

    -Si no se encuentra ningún elemento igual a la clave,
    la búsqueda arroja el siguiente resultado:
        "No se ha encontrado ninguna coincidencia".

    Los parámetros de entrada de esta función, son:
        -arr: La lista de elementos.
        -x: El nº a encontrar.
        -N: El tamaño de la lista.
    Cuando la función termina encuentra el nº lo devuelve
    junto con el tiempo transcurrido. En caso de que el nº
    que se quiere buscar no existiese, entonces, la función
    devuelve "-1".
    """
    for i in range(0, N): 
        if (arr[i] == x): 
            return i 
    return -1


def figuras(lista_equi, times_bin, times_lin):
    """
    Esta función recibe como argumentos, todos los tiempos
    de ejecución, que son los siguientes:
        -times_lin: vector con los tiempos de las ejecuciones
        acometidas por el algoritmo `linear_search`
        -times_sent: vector con los tiempos de las ejecuciones
        acometidas por el algoritmo `sentinel_search`

    Nótese como esta función no devuelve ningún parámetro o
    valor, simplemente se limita a gráficar los tiempos de
    ejecución de las dos funciones.
    """
    # print(f'Este es el tamaño de times_lin: {len(times_lin)}')
    # print(f'Este es el tamaño de times_sent: {len(times_sent)}')

    # print(f'Este es el primer tamaño de times_lin: {len(times_lin[1])}')
    # print(f'Este es el primer tamaño de times_sent: {len(times_sent[1])}')

    # log_bin = [math.log(x) for x in times_bin]  # logarithmic
    # log_met = [math.log(x) for x in times_met]  # logarithmic
    # log_ter = [math.log(x) for x in times_ter]  # logarithmic

    plt.figure()
    plt.plot(lista_equi, times_bin, marker='o')
    plt.plot(lista_equi, times_lin, marker='o')
    plt.legend(["binary_search", "linear_search"])
    plt.xlabel('$n$')
    plt.ylabel('$t(n)$')
    # plt.yscale('log')
    # plt.xscale('log')
    plt.title("Tiempos de ejecución para los 2 tipos de búsqueda")
    plt.show()

    plt.figure()
    plt.plot(lista_equi, times_bin, marker='o')
    plt.plot(lista_equi, times_lin, marker='o')
    plt.legend(["binary_search", "linear_search"])
    plt.xlabel('$n$')
    plt.ylabel('$t(n)$')
    plt.yscale('log')
    plt.xscale('log')
    plt.title("Tiempos de ejecución para los 2 tipos de búsqueda [log(·)]")
    plt.show()


def main():
    # Creamos la lista de los 100 elems equiespaciados
    lista_equi_1 = [10, 100, 1000, 10000, 100000, 1000000]
    lista_equi_2 = [4**x for x in range(10)]
    listas = [lista_equi_1, lista_equi_2]

    for lista in listas:
        print(f"Lista generada: \n {lista}\n")

        # Ahora se definen los tiempos
        times_bin = []
        times_lin = []
        for i in lista:
            lista_i = lst_bloques[:i]
            # print(f'Esta es la lista {i}\n{lista_i}')
            tam = len(lista_i)

            last_el = lista_i[-1]
            print(f'Hay que encontrar el siguiente nº: {last_el}')

            # Ahora llamamos a las tres funciones
            start_time1 = time.time()
            pos_bin = binary_search(lista_i, 0, tam-1, last_el)
            start_time2 = time.time()
            pos_lin = linear_search(lista_i, tam-1, last_el)
            stop_time1 = time.time()

            # Ahora calculamos los tiempos para cada función
            bin_time = start_time2 - start_time1
            lin_time = stop_time1 - start_time2

            # Comprobamos que se han encontrado los números correctos
            print(f"bin ha encontrado: {lista_i[pos_bin]} que coincide con {last_el}")
            print(f"lin ha encontrado: {lista_i[pos_lin]} que coincide con {last_el}\n")

            # Agregamos los tiempos de exec para cada tamaño, cuidado con los 0's
            times_bin.append(bin_time)
            times_lin.append(lin_time)

        print(f'Los tiempos de binary_search(): \n{times_bin}\n')
        print(f'Los tiempos de meta_search(): \n{times_lin}\n')

        # representamos entonces
        figuras(lista, times_bin, times_lin)


def temporitzadors():
    """
    En esta función se hace uso de timeit.Timer, a fin de poder 
    conocer más en detalle los tiempos de ambas ejecuciones.
    Esta librería nos permite lanzar los dos algoritmos varias 
    veces y además repetir una misma ejecución varias veces.
    """
    # Creamos la lista de los 100 elems equiespaciados
    lista_equi = [10, 100, 1000, 10000, 100000, 1000000]
    print(f"Lista generada: \n {lista_equi}\n")

    # Ahora se definen los tiempos
    times_bin = []
    times_lin = []
    for i in lista_equi:
        lista_i = lst_bloques[:i]
        # print(f'Esta es la lista {i}\n{lista_i}')
        tam = len(lista_i)

        last_el = lista_i[-1]
        print(f'\nHay que encontrar el siguiente nº: {last_el}')

        # ahora analizamos los tiempos con timeit
        # definim primer el Timer:
        t_bin = timeit.Timer(lambda: binary_search(lista_i, 0, tam-1, last_el))
        t_lin = timeit.Timer(lambda: linear_search(lista_i, tam-1, last_el))

        # si volem executar aquesta instrucció 10 vegades:
        #print(t.timeit(number=10))
        num = 50
        rep = 10

        # si volem executar la instrucció 10 vegades, i repetir el procés 5 vegades:
        tt_bin = t_bin.repeat(number=num, repeat=rep)
        times_bin.append(tt_bin)

        tt_lin = t_lin.repeat(number=num, repeat=rep)
        times_bin.append(tt_lin)

        print(tt_bin)
        print(tt_lin)

        # com es comenta a la documentació de Python:
        # https://docs.python.org/3/library/timeit.html
        # "Note It’s tempting to calculate mean and standard deviation from the result vector and report these.
        # However, this is not very useful. In a typical case, the lowest value gives a lower bound
        # for how fast your machine can run the given code snippet; higher values in the result vector are
        # typically not caused by variability in Python’s speed, but by other processes interfering
        # with your timing accuracy. So the min() of the result is probably the only number you should
        # be interested in. After that, you should look at the entire vector and apply common sense
        #rather than statistics.

        # Per tant, el millor que podem fer amb aquests 5 valors és agafar el valor mínim:
        print('\nEl valor min de binary_search: ',
              min(t_bin.repeat(number=num, repeat=rep)))

        print('El valor min de linear_search: ',
              min(t_lin.repeat(number=num, repeat=rep)))

    # representamos entonces
    figuras([x for x in range(1,rep+1)], tt_bin, tt_lin)


lst_bloques = sorted(lst_bloques)
main()
temporitzadors()
```

Implementa los tres primeros algoritmos de la lista: Binary Search, Meta Binary Search y Ternary Search. Ponlos en una misma gráfica para observar si hay alguna diferencia significativa. Recuerda que estos algoritmos son de tipo log(N), y por tanto si pones escala logarítmica en el tiempo de ejecución tendrías que ver un comportamiento lineal.

**Solución 4.2**

Nótese que para la implementación de estos tres algoritmos, se ha optado por las variantes iterativas de los algoritmos, a fin de poder medir más fácilmente su tiempo de ejecución, ya que se llama una vez a cada función, en vez de varias veces como se hace en las variantes recursivas.

Véase a continuación la implementación de estos tres algoritmos en la siguiente celda de código.

```python
def binary_search(arr, le, r, x):  # iterative approach
    """
    La búsqueda binaria se define como un algoritmo de búsqueda
    utilizado en una matriz ordenada dividiendo repetidamente el
    intervalo de búsqueda por la mitad. La idea de la búsqueda
    binaria es utilizar la información de que la matriz está ordenada
    y reducir la complejidad temporal a O(log N).

    Parámetros de entrada:
        -arr: es la lista donde se va a llevar a cabo la búsqueda.
        -le: es el índice donde se va a comenzar a buscar el nº.
        -r: es el tamaño de la lista.
        -x: es el nº que se quiere buscar en la lista.
    """
    while le <= r:

        mid = le + (r - le) // 2

        # Check if x is present at mid
        if arr[mid] == x:
            return mid

        # If x is greater, ignore left half
        elif arr[mid] < x:
            le = mid + 1

        # If x is smaller, ignore right half
        else:
            r = mid - 1

    # If we reach here, then the element
    # was not present
    return -1


def meta_binary_search(A, key_to_search):
    """
    La metabúsqueda binaria, también conocida como búsqueda binaria
    unilateral, es una variación del algoritmo de búsqueda binaria
    que se utiliza para buscar en una lista o matriz ordenada de
    elementos. Este algoritmo está diseñado para reducir el número
    de comparaciones necesarias para buscar un elemento determinado
    en la lista.

    Parámetros de entrada:
        -A: lista donde se va a buscar el número
        -key_to_search: nº que se quiere buscar en la lista A.

    Valor de retorno:
        -pos: posición donde se encuentra el nº en la lista.
    """
    n = len(A)
    # Set number of bits to represent
    lg = int(math.log2(n-1)) + 1

    pos = 0
    for i in range(lg - 1, -1, -1) :
        if (A[pos] == key_to_search):
            return pos

        # Incrementally construct the
        # index of the target value
        new_pos = pos | (1 << i)

        # find the element in one
        # direction and update position
        if ((new_pos < n) and
            (A[new_pos] <= key_to_search)):
            pos = new_pos

    # if element found return
    # pos otherwise -1
    return (pos if(A[pos] == key_to_search) else -1)


def ternary_search(l, r, key, ar): # iterative approach
    """
    La búsqueda ternaria es un algoritmo de búsqueda que se utiliza
    para encontrar la posición de un valor objetivo dentro de una
    matriz ordenada. Se basa en el principio de dividir la matriz 
    en tres partes en lugar de dos, como en la búsqueda binaria.

    Parámetros de entrada:
        -l: indice de la lista donde se comienza a buscar.
        -r: tamaño de la lista.
        -key: nº que se quiere buscar en la lista.
        -ar: lista en la que se va a llevar a cabo la búsqueda
    Valores de retorno:
        -mid: en caso de encontrar el nº deseado, se devuelve la
            posición del mismo en la lista.
        -(-1): en caso de no encontrar el nº deseado, se devuelve
            un -1.
    """
    while r >= l:

        # Find mid1 and mid2
        mid1 = l + (r-l) // 3
        mid2 = r - (r-l) // 3

        # Check if key is at any mid
        if key == ar[mid1]:
            return mid1
        if key == ar[mid2]:
            return mid2

        # Since key is not present at mid,
        # Check in which region it is present
        # Then repeat the search operation in that region
        if key < ar[mid1]:
            # key lies between l and mid1
            r = mid1 - 1
        elif key > ar[mid2]:
            # key lies between mid2 and r
            l = mid2 + 1
        else:
            # key lies between mid1 and mid2
            l = mid1 + 1
            r = mid2 - 1

    # key not found
    return -1


def figuras(lista_equi, times_bin, times_met, times_ter):
    """
    Esta función recibe como argumentos, todos los tiempos
    de ejecución, que son los siguientes:
        -times_lin: vector con los tiempos de las ejecuciones
        acometidas por el algoritmo `linear_search`
        -times_sent: vector con los tiempos de las ejecuciones
        acometidas por el algoritmo `sentinel_search`

    Nótese como esta función no devuelve ningún parámetro o
    valor, simplemente se limita a gráficar los tiempos de
    ejecución de las dos funciones.
    """
    # print(f'Este es el tamaño de times_lin: {len(times_lin)}')
    # print(f'Este es el tamaño de times_sent: {len(times_sent)}')

    # print(f'Este es el primer tamaño de times_lin: {len(times_lin[1])}')
    # print(f'Este es el primer tamaño de times_sent: {len(times_sent[1])}')

    plt.figure()
    plt.plot(lista_equi, times_bin, marker='o')
    plt.plot(lista_equi, times_met, marker='o')
    plt.plot(lista_equi, times_ter, marker='o')
    plt.legend(["binary_search", "meta_binary_search", "ternary_search"])
    plt.xlabel('$n$')
    plt.ylabel('$t(n)$')
    plt.yscale('log')
    plt.xscale('log')
    plt.title("Tiempos de ejecución para los 3 tipos de búsqueda (log(·))")
    plt.show()


def main():
    """
    Función principal.
    """
    # Creamos la lista de los 100 elems equiespaciados
    lista_equi = [10, 100, 1000, 10000, 100000, 1000000]
    print(f"Lista generada: \n {lista_equi}\n")

    # Ahora se definen los tiempos
    times_bin = []
    times_met = []
    times_ter = []
    for i in lista_equi:
        lista_i = lst_bloques[:i]
        # print(f'Esta es la lista {i}\n{lista_i}')
        tam = len(lista_i)

        last_el = lista_i[-1]
        print(f'Hay que encontrar el siguiente nº: {last_el}')

        # Ahora llamamos a las tres funciones
        start_time1 = time.time()
        pos_bin = binary_search(lista_i, 0, tam-1, last_el)
        start_time2 = time.time()
        pos_meta = meta_binary_search(lista_i, last_el)
        stop_time1 = time.time()
        pos_tern = ternary_search(0, tam-1, last_el, lista_i)
        stop_time2 = time.time()

        # Ahora calculamos los tiempos para cada función
        bin_time = start_time2 - start_time1
        meta_time = stop_time1 - start_time2
        tern_time = stop_time2 - stop_time1
        
        # Comprobamos que se han encontrado los números correctos
        print(f"bin ha encontrado: {lista_i[pos_bin]} que coincide con {last_el}")
        print(f"meta ha encontrado: {lista_i[pos_meta]} que coincide con {last_el}")
        print(f"tern ha encontrado: {lista_i[pos_tern]} que coincide con {last_el}\n")

        # Agregamos los tiempos de exec para cada tamaño, cuidado con los 0's
        times_bin.append(bin_time)
        times_met.append(meta_time)
        times_ter.append(tern_time)

    print(f'Los tiempos de binary_search(): \n{times_bin}\n')
    print(f'Los tiempos de meta_search(): \n{times_met}\n')
    print(f'Los tiempos de tern_search(): \n{times_ter}')

    # representamos entonces
    figuras(lista_equi, times_bin, times_met, times_ter)


lst_bloques = sorted(lst_bloques)
main()
```

Para calcular esta mejora se han recogido los dos tiempos de ejecución, asociados a las ejecuciones de las funciones ejecutadas secuencialmente (`t_total`) y meidante *threads* (`t_total_threads`)

```python
def calculo_mejora(t_total, t_tot_threads):
    """
    Como bien indica el nombre de esta función,
    esta cálcula la mejoría en (%) del tiempo
    de ejecución para dos funciones ejecutadas
    secuencialmente y mediante threads:

    Parámetros de entrada:
        -t_total: es el tiempo que tarda en ejecutarse
            el programa secuencialmente.
        -t_total_threads: es el tiempo que tarda en ejecutarse
            el programa mediante la implementción de dos 
            threads
    Parámetros de salida:
        -mejora: valor numérico que indica la mejora porcentual
    """
    mejora = ((t_total - t_tot_threads) / t_total) * 100
    return mejora

mejora = calculo_mejora(t_total, t_tot_threads)
print(f'La mejora es de un {mejora} %')
```

### 6.5
Explica con tus palabras las ventajas que has obtenido en separar los dos procesos en diferentes threads.

**Solución 6.5**

Al haber creado dos hilos/*threads* de ejecución para cada una de las dos tareas (`compute_pi`, `descargar_PDFs`), hemos podido explotar los recursos de nuestra máquina local paralelamente (CPU + escrituras en disco). Ahora bien, esto ha sido posible principalmente gracias a la naturaleza de las dos funciones. Para poder analizar y entender, porque se ha podido mejorar el tiempo de ejecución, hay que tener claro que recursos físicos necesitan ambas funciones, para ejecutarse.

* `compute_pi`, es una función matemática, que como bien sabemos, se centra en el cálculo de decimales del número $\pi$. Para acometer este cálculo se ha implementado uno de los varios algoritmos disponibles: el algoritmo de Chudnovsky. Este algoritmo transforma en decimal varios números, luego implementa un bucle `for` que itera (n+3) (n: nº de decimales) y que va calculando de manera iterativa, numerosas sumas, restas, multiplicaciones y divisiones. Este tipo de operaciones aritméticas, requieren de la CPU para poder acometer los cálculos, por lo que si se quieren calcular 7000 decimales de $\pi$ la CPU estará muy ocupada durante un periodo de tiempo. 

* `descargar_PDFs`, es una función, que dada una lista de carácteres con las direcciones web de cada uno de los PDFs, descarga dichos documentos. Esta función solamente implementa un bucle a fin de poder descargar de manera iterativa los documentos deseados, pero no es una función que ejecte ningún cálculo de alto nivel, por lo que no hace el mismo uso de la CPU que la función `compute_pi`, de hecho hará muy poco uso de ella, pues lo único que hace es hacer un `request` de la url, y luego escribe en disco, para poder guardar el PDF descargado.

Por lo tanto, al haber separado las dos tareas en procesos diferentes, hemos permitido a nuestra máquina calcular los diferentes decimales de $\pi$ (usar la CPU) e ir descargando los diferentes PDFs (comprobar dirección url y escribir en disco) al mismo tiempo. Y al poder realizar varias tareas al mismo tiempo, que precisan de diferentes recursos, el tiempo de ejecución disminuye.

Es por esto, que si quisiésemos ejecutar el cálculo de 7000 decimales de $\pi$ en un *thread* y luego calcular 500 decimales de $\pi$ en otro *thread* no veríamos ninguna mejoría, porque ambas funciones harían uso los mismos módulos *hardware* de nuestro ordenador.

Aun así, si las tareas dependen en un grado muy alto de la CPU, entonces, el uso de procesos, podría mejorar los tiempos de ejecución considerablemente, pues se aprovecharían más los recursos del sistema. No osbtante, esto se ve en el siguiente ejercicio.

### 7.1
En un proceso calcularemos los 7000 decimales del número pi. Y también crearemos un proceso para cada descarga de fichero. Igual que antes, queremos comparar la versión con un solo proceso (ya calculado) con la versión multiproceso. Calcula el tiempo total.

**Solución 7.1**

Véase a continuación la implementación de este apartado.

```python
def compute_pi_mod(p_index, n):
    """
    This function calculates the value of pi to 'n' number of decimal places
    Args:
    t_index: indica el threada donde se va a computar
    n:   precision(Decimal places)
    Returns:
    pi:   the value of pi to n-decimal places
    """
    logging.info("[P{}]\tSe empieza con: {} decimales".format(p_index, n))
    decimal.getcontext().prec = n + 3
    decimal.getcontext().Emax = 999999999

    C = 426880 * decimal.Decimal(10005).sqrt()
    K = decimal.Decimal(6)
    M = decimal.Decimal(1)
    X = decimal.Decimal(1)
    L = decimal.Decimal(13591409)
    S = L

    # Para más precisión, calculamos hasta n+3
    # y truncamos los dos últimos digitos
    for i in range(1, n+3):
        M = decimal.Decimal(M * ((1728*i*i*i)-(2592*i*i)+(1104*i)-120)/(i*i*i))
        L = decimal.Decimal(545140134+L)
        X = decimal.Decimal(-262537412640768000*X)
        S += decimal.Decimal((M*L) / X)

    decims = str(C/S)[:-2]
    logging.info("[P{}]\t {} decimales de pi calculados".format(p_index, n))
    return decims  # Pi is C/S


def descarga_PDFs_mod(p_index, url):
    """
    Esta función se encarga de descargar los PDFs asociados a
    cada una de las urls contenidas en la lista urls.

    Parámetros de entrada:
        -urls: nombres a modo de string,
            de cada una de las direcciones web de los pdfs.
    """
    logging.info("[P{}]\tSe empieza con: {}".format(p_index, url))
    r = requests.get(url, stream=True)
    # nos quedamos solo con el nombre del PDF
    nombre_pdf = os.path.basename(url)

    ruta = os.path.join(os.getcwd(), nombre_pdf)
    with open(ruta, 'wb') as pdf_object:
        pdf_object.write(r.content)
        print(f'Se ha guardado {nombre_pdf}')
        logging.info("[P{}]\tFin".format(p_index))
        # return True


def main():
    # introducimos en una lista las direcciones web de los PDFs
    lista_urls = ["https://www.cs.umb.edu/~offner/files/pi.pdf",
                  "https://www.davidhbailey.com/dhbpapers/pi-formulas.pdf",
                  "https://www.davidhbailey.com/dhbtalks/dhb-conant.pdf",
                  "https://arxiv.org/pdf/1912.05319.pdf",
                  "https://maa.org/sites/default/files/pdf/pubs/amm_supplements/Monthly_Reference_12.pdf"]

    # Creamos los 2 threads que ejecutan compute_pi() o
    # descarga_PDFs() y los iniciamos
    procesos = []

    for i in range(6):
        if i == 0:
            proceso = Process(target=compute_pi_mod, args=(i, 7000))
        else:
            proceso = Process(target=descarga_PDFs_mod, args=(i, lista_urls[i-1]))

        procesos.append(proceso)
        proceso.start()

    # Esperamos a que los 2 threads finalicen
    for i, proceso in enumerate(procesos):
        logging.info("[M]\tWaiting to join process {}".format(i))
        proceso.join()
        logging.info("[M]\tProcess {} joined!".format(i))

    logging.info("[M]\tDONE!")


start_time = time.time()
main()
stop_time = time.time()
t_tot_procesos = stop_time - start_time
print(f'El tiempo de ejecución es de {t_tot_procesos} segundos')
```
